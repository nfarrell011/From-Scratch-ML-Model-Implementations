### Import modules

```python
from sklearn.datasets import make_blobs
from kmeans_class import KMeans
import matplotlib.pyplot as plt
```

### Make some globular test data

```python
X, y  = make_blobs(n_samples = 500, n_features = 2, centers = 4, random_state = 42)
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:, 0], X[:, 1])
X.shape
```

### Instantiate model

This model is instaniated with a **bad** choice for $k$

```python
model = KMeans(data = X, max_iters = 10, random_seed = 42)
```

### Find the best k using elbow method

```python
model.find_best_k()
```

### Fit model with a poor choice for k; k = 7

```python
model.fit(k = 7)
```

### Inspect internal indices

```python
model.get_intertia()
sil_score = model.get_silhouette_score()
simp_sil_score = model.get_simplified_silhouette_score()
print("Interia:", model.inertia)
print("Silhouette Score:", sil_score)
print("Simplified Silhouette Score:", simp_sil_score)

# Cluster portion will only be visible for 2D data
model.plot_silhouette_scores()
```

### Fit model with k discovered using elbow method

```python
model = KMeans(data = X, max_iters = 5, random_seed = 42)
model.fit(k = 4)
model.get_intertia()
sil_score = model.get_silhouette_score()
simp_sil_score = model.get_simplified_silhouette_score()
print("Interia:", model.inertia)
print("Silhouette Score:", sil_score)
print("Simplified Silhouette Score:", simp_sil_score)
model.plot_silhouette_scores()
```
